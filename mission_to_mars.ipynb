```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
#!pip install splinter
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
scrape_site = "https://redplanetscience.com/"
```

```python
browser.visit(scrape_site)
```

```python
#Html object
html = browser.html

soup = bs (html, 'html.parser')
display (soup)
```

```python
find_title = soup.find_all ("div", class_="content_title")
find_title
```

```python
find_title = find_title[3].text
find_title
```

```python
find_para = soup.find_all ("div", class_="article_teaser_body")
find_para
```

```python
find_para = find_para[3].text
find_para
```
